```python
import pandas as pd
from datetime import datetime

# Load data
stops = pd.read_csv('Datasets/stops.txt')
trips = pd.read_csv('Datasets/trips.txt')
stop_times = pd.read_csv('Datasets/stop_times.txt')  # your stop_times data
vehicles = pd.read_csv('data/bus_data_current_chicago.csv')       # your bus tracker data

# Step 1: Link tatripid → trip_id via trips.schd_trip_id
trips['schd_trip_id'] = trips['schd_trip_id'].astype(str)
vehicles['tatripid'] = vehicles['tatripid'].astype(str)
trips['schd_trip_id_stripped'] = trips['schd_trip_id'].str.replace('020$', '', regex=True).str.lstrip('0')
# Then try merging on that vs tatripid

vehicles = vehicles.merge(
    trips[['trip_id', 'schd_trip_id', 'route_id']],
    left_on='tatripid',
    right_on='schd_trip_id',
    how='left'
)

# Step 2: Join stop_times to get scheduled stops for each trip
stop_times = stop_times.merge(stops[['stop_id','stop_name','stop_lat','stop_lon']], on='stop_id')

# Step 3: For each vehicle, find the NEXT scheduled stop
# (the first stop where shape_dist_traveled > vehicle's pdist)
def find_next_stop(row, st):
    trip_stops = st[st['trip_id'] == row['trip_id']].sort_values('shape_dist_traveled')
    next_stops = trip_stops[trip_stops['shape_dist_traveled'] >= row['pdist']]
    if next_stops.empty:
        return None
    return next_stops.iloc[0]

# Step 4: Compare scheduled arrival time vs current time
# Parse vehicle timestamp
vehicles['tmstmp'] = pd.to_datetime(vehicles['tmstmp'], format='%Y%m%d %H:%M')

# For each vehicle, get scheduled time at next stop and compute delay
# stop_times arrival_time is like "11:37:00" — convert to today's datetime for comparison
def get_delay_minutes(vehicle_time, scheduled_time_str):
    h, m, s = map(int, scheduled_time_str.split(':'))
    scheduled = vehicle_time.replace(hour=h, minute=m, second=s)
    return (vehicle_time - scheduled).total_seconds() / 60  # positive = late
```
